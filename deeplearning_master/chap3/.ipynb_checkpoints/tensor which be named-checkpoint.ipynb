```python
import torch
```

```python
img_t = torch.randn(3, 5, 5) #채널 크기, 행 크기, 열
weights = torch.tensor([0.2126, 0.7152, 0.0722])
print("img_t\n",img_t)
print("mean of img_t channal \n",img_t.mean(-3)) # 같은 위치의 값 in tensor 들의 평균
```

```python
batch_t1 = torch.randn(2, 3, 5, 5) # 배치 크기 , 채널크기, 행 크기, 열크기
batch_t2 = torch.randn(3, 3, 5, 5)
print(batch_t1)
print()
print(batch_t2)
```

```python
print("0 번째 차원은 배치, 여기서는 0 번째 배치 (3개의 채널이 5행 5열을 이룸)\n", batch_t1[0])
print()
print("1 번째 차원은 해당 배치에서 채널의 값, 여기서는 0번 째 배치의 0번째 채널 값\n", batch_t1[0,0])
```

```python
batch_t = torch.randn(2, 3, 5, 5)
img_gray_naive = img_t.mean(-3)
batch_gray_naive = batch_t.mean(-3)
print(f'img_gray shape: {img_gray_naive.shape}\nbatch_gray_naive shape: {batch_gray_naive.shape}')
```

```python
unsqueezed_weights = weights.unsqueeze(-1).unsqueeze_(-1) 
# unsqueeze(-1)는 크기가 1인 차원을 마지막 차원에 추가
# _를 함수 뒤에 붙이면 호출한 텐서 자체를 변경하나 여기서는 반환 된 값을 바꿨기에 원본은 바뀌지 않는다.
img_weights = (img_t * unsqueezed_weights)
batch_weights = (batch_t * unsqueezed_weights)
img_gray_weighted = img_weights.sum(-3)
batch_gray_weighted = batch_weights.sum(-3)
print("batch_weights",batch_weights.shape)
print("batch_t",batch_t.shape)
print("unsqueezed_weights",unsqueezed_weights.shape)
```

```python
img_gray_weighted_fancy = torch.einsum('...chw,c->...hw', img_t, weights) # 효율성을 위한 einsum합, 이런게 있다는 정도만 숙지
batch_gray_weighted_fancy = torch.einsum('...shw,c->...hw', batch_t, weights)
batch_gray_weighted_fancy.shape
```

```python
weights_named = torch.tensor([0.2126, 0.7152, 0.0722], names=['channels']) # 텐서에 이름 지정
weights_named
# 경고: 현재 텐서에 이름을 지정하는 기능은 실험 단계에 있습니다 (2024.02.02)
```

```python
img_named = img_t.refine_names(..., 'channels', 'rows', 'columns') # refine_names : 텐서 먼저 만들고 나중에 이름 지정
batch_named = batch_t.refine_names(..., 'channels', 'rows', 'columns') # ...를 사용하면 명시한 차원 이외의 차원의 이름들은 건드리지 않는다.
print("img named:", img_named.shape, img_named.names)
print("batch named:", batch_named.shape, batch_named.names)
# rename을 이용하면 이미 있는 이름을 바꿀 수 있다
# 지우려면 이름으로 None을 넘긴다
```

```python
weights_aligned = weights_named.align_as(img_named) 
# align_as : 빠진 차원을 채워주고 존재하는 차원을 올바른 순서로
# 여기서는 weights_named의 텐서를 img_named의 텐서와 동일한 차원을 갖게 만들어 
weights_aligned.shape, weights_aligned.names
```

```python
gray_named = (img_named * weights_aligned).sum('channels')
gray_named.shape, gray_named.names
```

```python
print("img_named\n", img_named.shape, img_named.names)
print("\nweights_aligned\n", weights_aligned.shape, weights_aligned.names)
print("\nimg_named\n", img_named)
print("\nweights_aligned\n", weights_aligned)
print("\nimg_named[..., :3] \n",img_named[..., :3])
gray_named = (img_named[..., :3] * weights_aligned).sum('channels')
# ...은 모든 남은 차원을 의미, :3은 처음부터 3번 인덱스를 의미 
# 여기서는 img_named의 마지막 차원에서 3개의 요소와 나머지 차원들 * weights_aligned (브로드 캐스팅 되서 weights_aligned가 [3, 5, 3]로 됨
print("\ngray_named\n",gray_named)
```

```python
gray_plain = gray_named.rename(None) # 이름 있는 텐서를 사용하는 연산을 함수 밖에서도 사용 가능하게 함
gray_plain.shape, gray_plain.names
```
