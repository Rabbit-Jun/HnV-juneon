1에서 100까지 출력
3의 배수는 Fizz 출력
5의 배수는 Buzz 출력
3과 5의 공배수는 FizzBuzz 출력

```python
for i in range(1,101):
    if i%3==0 and i %5==0:
        print('FizzBuzz')
    elif i %3 ==0:
        print('Fizz')
    elif i%5==0:
        print('Buzz')
    
    else:
        print(i)
```
