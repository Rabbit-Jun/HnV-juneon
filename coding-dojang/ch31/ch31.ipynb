```python
def hello():
    print('hello,world!')
    hello() #재귀 호출 (자기자신을 계속 호출하다가 파이썬에서 정한 최대 재귀 깊이인 1,000을 초과하면  RecursionError )
 
hello() 
```

```python
def hello(count):
    if count == 0: #종료 조건을 만듦. count가 0이면 다시 hello함수를 호출하지 않고 끝냄
        return
    
    print('Hello,world!', count)

    count -=1   # count를 1 감소시킨 뒤
    hello(count)    # 다시 hello에 넣음

hello(5)    #hello 함수 호출

```

```python
def factorial(n):
    if n ==1: #n이 1일 때
        return 1 #1을 반환하고 재귀호출을 끝냄
    return n * factorial(n-1) #n과 factorial함수에 n-1을 넣어서 반환된 값을 곱함

print(factorial(5))
```

```python
def is_palindrome(word):
    a= False
    if word ==word[::-1]:
        a= True
    return a
print(is_palindrome('hello'))
print(is_palindrome('level'))
```
