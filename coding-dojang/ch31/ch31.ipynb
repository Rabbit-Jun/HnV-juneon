```python
def factorial(n):
    if n ==1: #n 1 
        return 1 #1   
    return n * factorial(n-1) #n factorial n-1    

print(factorial(5))
```

```python
def fib(n):
    if n ==1:
        return 1
    elif n ==0:
        return 0
    return fib(n-1) +fib(n-2)

n = int(input())
print(fib(n))
```
