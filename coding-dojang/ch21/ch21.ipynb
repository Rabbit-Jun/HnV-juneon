```python
import turtle as t
t.shape('turtle')
for _ in range(3):
    t.forward(100) #100픽셀만큼 앞으로 이동
    t.right(90) #오른쪽으로 90도 회전
t.fd(100) #forward
t.rt(90) #right
t.bk(50) #back
t.lt(60) # left
t.mainloop()
```

```python
import turtle as t

n =6
t.shape('turtle')
t.color('orange') #펜 색 설정
t.begin_fill() #색칠할 영역 시작
for i in range(n):
    t.forward(100)
    t.right(360/n)
    
t.end_fill() #색칠할 영역 끝
```

```python
import turtle as t

n =60
t.shape('turtle')
t.speed('fastest') #거북이 속도 지정
for i in range(n):
    t.circle(120)
    t.right(360/n)
```
